```python
import numpy as np
import math
np.random.seed(42)
```

```python
nominal = np.ones((9,9)) * -1
```

```python
nominal[1][0] = 10
nominal[0] = -1
nominal
```

```python
avg = np.mean(nominal)
std = np.std(nominal)
max_n = np.max(nominal)
min_n = np.min(nominal)
avg, std, max_n, min_n
```

```python
scaler = abs(nominal - avg) /(max_n - min_n)
scaler
```

```python
learned = np.random.rand(9,9) * (-50)
max_l = np.max(learned)
min_l = np.min(learned)
print(learned)
print(np.exp(np.ravel(abs(learned - avg) /(max_l - min_l)))/np.sum(np.exp(np.ravel(abs(learned - avg) /(max_l - min_l)))))
```
